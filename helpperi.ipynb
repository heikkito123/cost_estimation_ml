```python
def get_totals(ws, ranges):
    """
    Get m2-totals from a worksheet. 
    Input: workbook['sheetname], list[ranges(first col, last col, first row, last row)]
    """
    totals = []

    for min_col, max_col, min_row, max_row in ranges:
        range_total = [
            cell
            for row in ws.iter_rows(
                min_col=min_col,
                max_col=max_col,
                min_row=min_row,
                max_row=max_row,
                values_only=True
            )
            for cell in row
            if cell is not None
        ]

        totals.extend(range_total)

    return sum(totals)
```
